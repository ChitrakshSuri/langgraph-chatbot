```python
from langgraph.graph import StateGraph, START, END
from langchain_openai import ChatOpenAI
from typing import TypedDict
from dotenv import load_dotenv
```

```python
load_dotenv()
```

```python
model = ChatOpenAI()
```

```python
class BlogState(TypedDict):
    title: str
    outline: str
    content: str
```

```python
def create_outline(state: BlogState) -> BlogState:
    title = state['title']

    prompt = f'Generate a detailed outline for a blog post with title {title}'

    outline = model.invoke(prompt).content

    state['outline'] = outline

    return state
```

```python
def create_blog(state: BlogState) -> BlogState:
    title = state['title']
    outline = state['outline']

    prompt = f'Generate a blog post on {title} with outline {outline}'

    blog = model.invoke(prompt).content

    state['content'] = blog

    return state
```

```python
# define graph
graph = StateGraph(BlogState)

# add node
graph.add_node('create_outline', create_outline)
graph.add_node('create_blog', create_blog)

# add edge
graph.add_edge(START, 'create_outline')
graph.add_edge('create_outline', 'create_blog')
graph.add_edge('create_blog', END)

# compile
workflow = graph.compile()
```

```python
initial_state = {'title': 'salary difference between an indian and an us software dev if indian dev salry is 60k in noida should be his salary in Ohio USA with same life style'}
final_state = workflow.invoke(initial_state)
print(final_state['content'])
```
